```python
import pandas as pd
import random

random.seed(42)

case_types = ["criminal", "civil", "administrative"]
provinces  = ["Beijing", "Guangdong", "Sichuan", "Xinjiang", "Shanghai"]
demo_texts = [
    "被告人认罪态度良好，系初犯，依法从轻处罚，判处有期徒刑一年，缓期执行。",
    "被告人犯罪情节严重，依法严厉打击，维护社会秩序，判处有期徒刑十年。",
    "证据不足，无法认定被告人有罪，依法宣告无罪。",
    "被告人违法行为恶劣，社会危害性大，从严惩处，判处有期徒刑五年。",
    "被告人认罪悔罪，积极赔偿损失，酌情从轻，判处拘役三个月。",
]

df = pd.DataFrame({
    "text":        [random.choice(demo_texts) for _ in range(50)],
    "case_type":   [random.choice(case_types)  for _ in range(50)],
    "province":    [random.choice(provinces)   for _ in range(50)],
    "year":        [random.randint(2015, 2023) for _ in range(50)],
    "court_level": [random.choice(["district", "intermediate", "high"]) for _ in range(50)],
})

print(f"Created {len(df)} demo verdicts")
print(df.head())
```

```python
print("Verdicts by case type:")
print(df["case_type"].value_counts())

print("\nVerdicts by province:")
print(df["province"].value_counts())

print("\nYear range:")
print(f"Earliest: {df['year'].min()}, Latest: {df['year'].max()}")
```
